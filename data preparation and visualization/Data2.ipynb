```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
import os
from tqdm import tqdm
from seaborn import set_style

set_style("whitegrid")
#Import the csv file
dfhl=pd.read_csv('../data/dfhl.csv')
dfhl.time=pd.to_datetime(dfhl.time)
dfhl.open=dfhl.open.astype(float);dfhl.volume=dfhl.volume.astype(float);dfhl.MACD=dfhl.MACD.astype(float)
dfhl.MACD_Hist=dfhl.MACD_Hist.astype(float);dfhl.MACD_Signal=dfhl.MACD_Signal.astype(float)
dfhl
```

```python
#make two-hour window for 'open'
tdata=dfhl.copy()
open_price=[]
for t in range(120):
    column=[]
    for i in range(20):
        td = pd.Timedelta(i, "d")
        column.extend(tdata.loc[(tdata.time>=pd.Timestamp(2023,4,28)+td) &
                                                (tdata.time<pd.Timestamp(2023,4,28,15)+td)].open.shift(t).values[::-1][:-119])
    tdf=pd.DataFrame(columns=[f'open {t}'])
    tdf[f'open {t}']=column;open_price.append(tdf)
dfopen=pd.concat(open_price,axis=1)
dfopen
```

```python
#make two-hour window for 'volume'
volume=[]
for t in range(120):
    column=[]
    for i in range(20):
        td = pd.Timedelta(i, "d")
        column.extend(tdata.loc[(tdata.time>=pd.Timestamp(2023,4,28)+td) &
                                                (tdata.time<pd.Timestamp(2023,4,28,15)+td)].volume.shift(t).values[::-1][:-119])
    tdf=pd.DataFrame(columns=[f'volume {t}'])
    tdf[f'volume {t}']=column;volume.append(tdf)
dfvolume=pd.concat(volume,axis=1)
dfvolume
```

```python
#make two-hour window for 'MACD'
MACD=[]
for t in range(120):
    column=[]
    for i in range(20):
        td = pd.Timedelta(i, "d")
        column.extend(tdata.loc[(tdata.time>=pd.Timestamp(2023,4,28)+td) &
                                                (tdata.time<pd.Timestamp(2023,4,28,15)+td)].MACD.shift(t).values[::-1][:-119])
    tdf=pd.DataFrame(columns=[f'MACD {t}'])
    tdf[f'MACD {t}']=column;MACD.append(tdf)
dfMACD=pd.concat(MACD,axis=1)
dfMACD
```

```python
#make two-hour window for 'Hist'
Hist=[]
for t in range(120):
    column=[]
    for i in range(20):
        td = pd.Timedelta(i, "d")
        column.extend(tdata.loc[(tdata.time>=pd.Timestamp(2023,4,28)+td) &
                                            (tdata.time<pd.Timestamp(2023,4,28,15)+td)].MACD_Hist.shift(t).values[::-1][:-119])
    tdf=pd.DataFrame(columns=[f'Hist {t}'])
    tdf[f'Hist {t}']=column;Hist.append(tdf)
dfHist=pd.concat(Hist,axis=1)
dfHist
```

```python
#make two-hour window for 'Signal'
Signal=[]
for t in range(120):
    column=[]
    for i in range(20):
        td = pd.Timedelta(i, "d")
        column.extend(tdata.loc[(tdata.time>=pd.Timestamp(2023,4,28)+td) &
                                        (tdata.time<pd.Timestamp(2023,4,28,15)+td)].MACD_Signal.shift(t).values[::-1][:-119])
    tdf=pd.DataFrame(columns=[f'Signal {t}'])
    tdf[f'Signal {t}']=column;Signal.append(tdf)
dfSignal=pd.concat(Signal,axis=1)
dfSignal
```

```python
#X_train
df=pd.concat([dfopen,dfvolume,dfMACD,dfHist,dfSignal],axis=1)
df
```
